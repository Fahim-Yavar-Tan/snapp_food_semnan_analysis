```python
import pandas as pd
from hazm import Normalizer, word_tokenize, stopwords_list, Lemmatizer
import re
```

```python
df = pd.read_csv('Snappfood-comments.csv')
```

```python
df
```

```python
text = df.loc[0, 'comment']
print('text\t\t', text)
normalizer = Normalizer()
normalized_text = normalizer.normalize(text)
print('normalized_text\t', normalized_text)
tokens = word_tokenize(normalized_text)
print('word_tokenize\t', tokens)
tokens = [token.lower() for token in tokens]
print('lower\t\t', tokens)
tokens = [re.sub(r'[^a-zA-Zآ-ی]', '', token) for token in tokens]
print('sub\t\t', tokens)
stop_words = set(stopwords_list())
tokens = [token for token in tokens if token not in stop_words]
print('stopword\t', tokens)
tokens = [token for token in tokens if token != '']
print('empty\t\t', tokens)
lemmatizer = Lemmatizer()
tokens_lemma = [lemmatizer.lemmatize(token) for token in tokens]
print('lemma\t\t', tokens_lemma)
```
